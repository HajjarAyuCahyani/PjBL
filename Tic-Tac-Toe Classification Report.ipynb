```python
from google.colab import files
uploaded = files.upload()
```

```python
import io
import pandas as pd
df = pd.read_csv(io.BytesIO(uploaded['tic-tac-toe.data']))
```

```python
df
```

```python
df.columns=[' top-left-square','top-middle-square','top-right-square','middle-left-square','middle-middle-square','middle-right-square','bottom-left-square','bottom-middle-square','bottom-right-square','Class']
df
```

```python
df.replace('x',1,inplace=True)
df.replace('o',2,inplace=True)
df.replace('b',3,inplace=True)
df.replace('positive',1,inplace=True)
df.replace('negative',0,inplace=True)
df.head()
```

```python
df.isnull().sum()
```

```python
# Menetapkan fitur (X) dan target (y)
X=df.drop(["Class"],axis = 1)
y=df["Class"]
```

# Train-Test Splitting

## Train-Test Split 70:30


```python
from sklearn.model_selection import train_test_split

# Split data menjadi data training dan testing
X_train7,X_test7, y_train7, y_test7 = train_test_split(X,y, test_size=0.3, random_state=1)

# Prediksi dan Evaluasi Performa Model
from sklearn.linear_model import LogisticRegression

# Membuat dan melatih model Regresi Logistik
model = LogisticRegression()
model.fit(X_train7, y_train7)

# Membuat prediksi pada data testing
split73 = model.predict(X_test7)

print('Jumlah data training:', len(X_train7))
print('Jumlah data testing:', len(X_test7))
```

## Train-Test Split 80:20

```python
from sklearn.model_selection import train_test_split

# Split data menjadi data training dan testing
X_train8,X_test8, y_train8, y_test8 = train_test_split(X,y, test_size=0.2, random_state=1)

# Prediksi dan Evaluasi Performa Model
from sklearn.linear_model import LogisticRegression

# Membuat dan melatih model Regresi Logistik
model = LogisticRegression()
model.fit(X_train8, y_train8)

# Membuat prediksi pada data testing
split82 = model.predict(X_test8)

print('Jumlah data training:', len(X_train8))
print('Jumlah data testing:', len(X_test8))
```

## Train-Test Split 90:10


```python
from sklearn.model_selection import train_test_split

# Split data menjadi data training dan testing
X_train9,X_test9, y_train9, y_test9 = train_test_split(X,y, test_size=0.1, random_state=1)

# Prediksi dan Evaluasi Performa Model
from sklearn.linear_model import LogisticRegression

# Membuat dan melatih model Regresi Logistik
model = LogisticRegression()
model.fit(X_train9, y_train9)

# Membuat prediksi pada data testing
split91 = model.predict(X_test9)

print('Jumlah data training:', len(X_train9))
print('Jumlah data testing:', len(X_test9))
```

```python
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix
df73 = confusion_matrix(y_test7, split73)
print("confusion matrix split data 70:30 \n", df73)
df82 = confusion_matrix(y_test8, split82)
print("confusion matrix split data 80:20 \n", df82)
df91 = confusion_matrix(y_test9, split91)
print("confusion matrix split data 90:10 \n", df91)
```

```python
report73 = classification_report(y_test7, split73)
print("Classification Report 70:30 =\n", report73)
report82 = classification_report(y_test8, split82)
print("Classification Report 80:20 =\n", report82)
report91 = classification_report(y_test9, split91)
print("Classification Report 90:10 =\n", report91)
```

# Cross Validation KFold

## Shuffle = False

```python
# cross validation
from sklearn.model_selection import KFold
n_folds = 4
accuracies=[]

kf = KFold(n_splits=n_folds)

for fold, (train_indeks, test_indeks) in enumerate(kf.split(df)):

  #Mendapatkan training dan testing pada fold
  train_data = df.iloc[train_indeks]
  test_data = df.iloc[test_indeks]

  #Memisahkan target dan fitur
  X_train = train_data.drop('Class', axis=1) #fitur
  X_test = test_data.drop('Class', axis=1) #fitur
  y_train = train_data['Class'] #target
  y_test = test_data['Class'] #target

  # Membuat dan melatih model Regresi Logistik
  model = LogisticRegression()
  model.fit(X_train, y_train)

  # Membuat prediksi pada data testing
  split = model.predict(X_test)

  cm= confusion_matrix(y_test, split)
  y_pred = model.predict(X_test)

  accuracy = accuracy_score(y_test, y_pred)
  accuracies.append(accuracy)
  #Menghitung persentase target pada masing-masing fold
  print("Fold number: ", fold)
  print("Original")
  print(df['Class'].value_counts().div(len(df)).mul(100)) # proporsi data original
  print("Training")
  print(y_train.value_counts().div(len(y_train)).mul(100))
  print("Testing")
  print(y_test.value_counts().div(len(y_test)).mul(100))
  print("confusion matrix: \n", cm)
  print("-"*30)

mean_accuracy = sum(accuracies) / len(accuracies)
print(accuracies)
print(mean_accuracy)
```

### Interpretasi


> dari hasil yang ada, ditunjukkan bahwa proporsi pada fold 0, 1, dan 3 tidak seimbang. observasi untuk data pengujian hanya berasal dari kelas positif saja atau negatif saja, yang ditunjukkan oleh persentase dari data testing sebesar 100% dari satu kelas. Sehingga, nilai akurasi yang didapat tergolong sangat rendah


> Berdasarkan hasil confusion matrix juga didapat bahwa nilai TN selalu 0, yang berarti model pada tiap fold sama sekali tidak dapat memprediksi kelas negatif dengan benar. prediksi yang dihasilkan oleh model pada kelas negatif menghasilkan kesalahan klasifikasi. selain itu, model secara rata-rata hanya memprediksi kelas positif saja terlepas dari benar atau salahnya prediksi.





## Shuffle = True



> Dikarenakan pembagian data test dan train yang tidak seimbang antara kelas positif dan negatif, maka dilakukan percobaan untuk mengacak data sebelum membagi menjadi 4 fold. Dengan harapan dapat membagi data test maupun data train secara merata



```python
# cross validation
from sklearn.model_selection import KFold
n_folds = 4
accuracies=[]
kf = KFold(n_splits=n_folds, shuffle=True)

for fold, (train_indeks, test_indeks) in enumerate(kf.split(df)):

  #Mendapatkan training dan testing pada fold
  train_data = df.iloc[train_indeks]
  test_data = df.iloc[test_indeks]

  #Memisahkan target dan fitur
  X_train = train_data.drop('Class', axis=1) #fitur
  X_test = test_data.drop('Class', axis=1) #fitur
  y_train = train_data['Class'] #target
  y_test = test_data['Class'] #target

  # Membuat dan melatih model Regresi Logistik
  model = LogisticRegression()
  model.fit(X_train, y_train)

  # Membuat prediksi pada data testing
  split = model.predict(X_test)

  cm= confusion_matrix(y_test, split)
  y_pred = model.predict(X_test)

  accuracy = accuracy_score(y_test, y_pred)
  accuracies.append(accuracy)

  #Menghitung persentase target pada masing-masing fold
  print("Fold number: ", fold)
  print("Original")
  print(df['Class'].value_counts().div(len(df)).mul(100)) # proporsi data original
  print("Training")
  print(y_train.value_counts().div(len(y_train)).mul(100))
  print("Testing")
  print(y_test.value_counts().div(len(y_test)).mul(100))
  print("confusion matrix: \n", cm)
  print("-"*30)

mean_accuracy = sum(accuracies) / len(accuracies)
print(accuracies)
print(mean_accuracy)
```

### Interpretasi
Dari di atas, dapat dilihat bahwa setelah dilakukan shuffle sebelum dibagi menjadi 4 fold, tidak ada ketimpangan yang berarti antara data train dan data test. Dengan demikian, nilai akurasi yang dihasilkan dapat meningkat dan mungkin sudah dapat dikatakan relatif baik. Kendati demikian, nilai rata-rata akurasi yang dihasilkan menunjukkan bahwa model tersebut masih dapat ditingkatkan lagi dalam menggeneralisasi data yang lebih baik.
